```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

A série temporal é da forma: $Z_t = \mu + ϵ_t$

Onde,

$\mu$: é a média constante

$ϵ$:  é a componente estocástica

A estimativa mais comum para μ é:

$\hat{\mu} = \bar{Z} = \frac{1}{N} ∑\limits_{t = 1}^{N} Z_t$



```python
t = np.arange(0,100,1)
n = len(t)

mu = 5
eps = np.zeros((n,1))

phi = 0.5

for i in range(1,n):
  eps[i] = phi * eps[i - 1] + random.gauss(0,1)


Zt = mu + eps
```

```python
plt.figure(figsize = (18,7))

plt.subplot(1,3,1)
plt.hlines(mu, xmin = 0, xmax = n)
plt.title('Média Constante - mu')

plt.subplot(1,3,2)
plt.plot(eps)
plt.title('Componente Estocática')

plt.subplot(1,3,3)
plt.plot(Zt)
plt.title('Série Temporal')
```

```python
estimativa_mu = 1/n * np.sum(Zt)

print('Estimativa para mu:',estimativa_mu)
```

```python
res = Zt - estimativa_mu


plt.figure(figsize = (18,7))
plt.plot(t,Zt, label = 'Série Temporal')
plt.hlines(estimativa_mu, color = 'r', xmin = 0, xmax = n, label = 'Estimativa de mu')
plt.plot(t,res,color = 'green', label = 'Resíduos')

plt.legend()
plt.show()
```
