```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/content/drive/MyDrive/Time Series Studies/IPI.XLS')
df['mês/ano'] = pd.to_datetime(df['mês/ano'])
```

```python
data_inicial = pd.to_datetime('1993-01-01')
data_final = pd.to_datetime('1996-12-31')

df2 = df[(df['mês/ano'] >= data_inicial) & (df['mês/ano'] <= data_final)].reset_index(drop='True')
```

```python
plt.figure(figsize = (12,7))
df2['ipialiment'].plot()
```

## representação Sazonal

```python
df2['Year'] = df2['mês/ano'].dt.year
df2['Month'] = df2['mês/ano'].dt.month
```

```python
df2['Year'].unique()
```

```python
fig, ax = plt.subplots(figsize = (18,7))  # Cria uma única figura e eixo

for ano in df2['Year'].unique():
  ax.plot('Month','ipialiment',marker = '*', data = df2[df2['Year'] == ano], label = f'Ano {ano}')

plt.title('IPI')
plt.legend()
plt.show()
```
