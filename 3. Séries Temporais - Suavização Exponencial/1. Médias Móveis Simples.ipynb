```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
```

A **média móvel simples** é uma técnica utilizada para estimar séries temporais cujo parâmetro $\mu(t)$ é constante ou varia lentamente ao longo do tempo. Consideremos uma série da forma:

$$Z(t) = \mu(t) + a(t)$$

onde $a(t)$ apresenta $E[a(t)] = 0$ e $Var[a(t)] = \sigma ^2_a$

A média móvel é calculada da seguinte maneira:

$$ M(t) = \frac{Z(t) + Z(t -1) + \dots + Z(t - r +1)}{r} $$

Onde $M(t)$ representa uma estimativa do nível $\mu(t)$ e $r$ define a janela de cálculo.

Essa média pode ser utilizada para previsão de valores futuros por meio da seguinte abordagem:

$$ \hat{Z}_t(h) = M(t), \enspace \forall h > 0$$

Ou seja, a previsão para os próximos instantes consiste em repetir a última média obtida.

Outra abordagem consiste em atualizar cada ponto futuro utilizando a estimativa anterior no cálculo da média móvel. No entanto, ao adotar esse método, é importante considerar que os **erros** das estimativas tendem a se **propagar ao longo do tempo**.




```python
df = pd.read_excel('/content/drive/MyDrive/Time Series Studies/poluicao.xls')
```

```python
plt.figure(figsize=(12,6))

plt.plot(df['DATA'],df['co'])

plt.title('Emissão de Gás Carbônico (CO) em SP')
plt.xlabel('Data')
```

```python
df['media_movel_co7']  = df['co'].rolling(window = 7 ).mean()
df['media_movel_co14'] = df['co'].rolling(window = 14).mean()
df['media_movel_co21'] = df['co'].rolling(window = 21).mean()
```

```python
plt.figure(figsize=(15,6))

plt.plot(df['DATA'],df['co'],label = 'Série Original', alpha = 0.5)
plt.plot(df['DATA'],df['media_movel_co7'] , color = 'red', linestyle ='dashed',  label = 'Média Móvel Simples (janela = 7)')
plt.plot(df['DATA'],df['media_movel_co14'], color = 'green', linestyle ='dotted',  label = 'Média Móvel Simples (janela = 14)')
plt.plot(df['DATA'],df['media_movel_co21'], color = 'purple', linestyle ='dashdot',  label = 'Média Móvel Simples (janela = 21)')

plt.title('Emissão de Gás Carbônico (CO) em SP')
plt.xlabel('Data')
plt.legend()
plt.tight_layout()
plt.show()
```
