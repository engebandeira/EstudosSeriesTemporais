```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Assim como a média móvel simples a suavização exponencial simples pode ser aplicada a séries com um comportamente constante ou que apresente uma baixa variação ao longo do tempo. Diferente da média simples que considera apenas um número $r$ de amostras anteriores, a suavização exponencial simples considera toda a série, atribuindo um maior peso as amostras mais próximas do ponto de referência.

Podemos calcular o valor suavisado a partir de:

$$ \bar{Z}(t) = \alpha Z(t) + (1 - \alpha) \bar{Z}(t -1) \quad \text{com} \quad
 \bar{Z}(0) = \bar{Z}(1) \enspace \text{e} \enspace t = 1,\dots ,N$$
Onde,

$\bar{Z}(t)$ é o valor exponencialmente suavizado e

$\alpha$ é a constante de suavização

Outra forma de expressar o cálculo é:

$$ \bar{Z}(t) = \alpha \sum_{ k = 0}^{t - 1} (1 - \alpha) Z(t)(t - k) + (1 - \alpha)^t \bar{Z}(0) $$

Quanto menor for o valor de $\alpha$ mais suave será a série suavizada resultante.



```python
df = pd.read_excel('/content/drive/MyDrive/Time Series Studies/poluicao.xls')
```

```python
df['exponential_005'] = df['no2'].ewm(alpha = 0.05, adjust=False).mean()
df['exponential_03'] = df['no2'].ewm(alpha = 0.3, adjust=False).mean()
```

```python
plt.figure(figsize=(15,6))

plt.plot(df['DATA'], df['no2'],label = 'Série Original', alpha = 0.6)
plt.plot(df['DATA'], df['exponential_005'], color = 'red'   , linestyle ='dashed', label = 'Suavização exponencial - alpha = 0.05')
plt.plot(df['DATA'], df['exponential_03'] , color = 'black' , linestyle ='dashdot', label = 'Suavização exponencial - alpha = 0.3')

plt.title('Emissão de Gás Carbônico (NO2) em SP')
plt.xlabel('Data')
plt.legend()
plt.tight_layout()
plt.show()
```
