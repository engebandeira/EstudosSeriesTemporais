```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,50,0.1)
len(t)
```

# TendÃªncia

```python
a, b = 9, 5.2
trend_linear = a + b * t # Linear trending

a, b = 9, 0.1
trend_exp = a * np.exp(b * t) # Exponential Trending


a, b = 9, 16
trend_log = a + b * np.log(t) # Logarithm Trending


plt.figure()

plt.plot(t, trend_log)

plt.title('Time Series Trending')

plt.tight_layout()
plt.show()
```

# Sazonalidade

```python
a, b = 12.5 , 15
freq_sin, freq_cos = 6, 12

sasonal_sin = a * np.sin( 2 * np.pi * t / freq_sin)
sasonal_cos = b * np.cos( 2 * np.pi * t / freq_cos)
sasonal = sasonal_sin + sasonal_cos

plt.figure(figsize=(15,6))

plt.plot(t, sasonal_sin, label = 'Sin')
plt.plot(t, sasonal_cos, label = 'Cos')
plt.plot(t, sasonal, label = 'Sin + Cos')
plt.title('Time Series Sasonality')
plt.legend()

plt.tight_layout()
plt.show()
```

## RuÃ­do Branco Uniforme

```python
n = len(t)
Ut = np.zeros((n,1))

for i in range(n):
    Ut[i] = random.uniform(0, 1)

plt.figure(figsize=(18,7))

plt.subplot(1,3,1)
plt.plot(t,Ut)
plt.title('RuÃ­do Branco Uniforme')

plt.subplot(1,3,2)
plt.hist(Ut, bins = 20)
plt.title('Histograma')

ax3 = plt.subplot(1,3,3)
plot_acf(Ut, lags = 30, title='FunÃ§Ã£o de AutocorrelaÃ§Ã£o Amostral', alpha = 0.05, ax = ax3)
plt.show()
```

# RuÃ­do Branco Gaussiano

```python
n = len(t)
Ut = np.zeros((n,1))

for i in range(n):
    Ut[i] = random.gauss(0, 1)

plt.figure(figsize=(18,7))

plt.subplot(1,3,1)
plt.plot(t,Ut)
plt.title('RuÃ­do Branco Uniforme')

plt.subplot(1,3,2)
plt.hist(Ut, bins = 20)
plt.title('Histograma')

ax3 = plt.subplot(1,3,3)
plot_acf(Ut, lags = 30, title='FunÃ§Ã£o de AutocorrelaÃ§Ã£o Amostral', alpha = 0.05, ax = ax3)
plt.show()
```
