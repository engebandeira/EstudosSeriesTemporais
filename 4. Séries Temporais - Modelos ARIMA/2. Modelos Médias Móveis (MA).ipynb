```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller
import random
```

```python
T = 2000      # Número de observações
mu = 0        # Média da série
theta1 = 0.8  # Coeficiente do termo MA(1)
sigma = 1     # Desvio padrão do ruído

# Gerando os termos de erro (ruído branco)
epsilon = np.random.normal(loc=0, scale=sigma, size=T)

# Inicializando a série
X = np.zeros(T)
q = 1

# Simulando o modelo MA(1)
for t in range(q, T):
    X[t] = mu + epsilon[t] - theta1 * epsilon[t-1]

# Plotando a série simulada
plt.figure(figsize = (18, 6))

plt.subplot(1,3,1)
plt.plot(X, label="Série MA(1)")
plt.axhline(y=0, color='r', linestyle='--', alpha=0.7)
plt.legend()
plt.title("Simulação de um Processo MA(1)")
plt.xlabel("Tempo")
plt.ylabel("Z_t")

plt.subplot(1,3,2)
plt.hist(X, bins = 50)
plt.title('Histograma')

ax3 = plt.subplot(1,3,3)
plot_acf(X, lags = 30, title = 'Função de Autocorrelação Amostral', alpha = 0.05, ax = ax3)
plt.ylim(ymin=-1.05,ymax=1.05)
```
