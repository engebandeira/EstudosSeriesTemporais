```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
import random
```

```python
model = ARIMA(endog=[0], order=(1, 1, 1))
simulated = model.simulate(params=[0.6, 0.5, 0.8], nsimulations=1500)

plt.plot(simulated)
plt.title("Simulação ARIMA(1,1,1) via statsmodels")
plt.grid(True)
plt.show()
```
